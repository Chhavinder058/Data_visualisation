# **Let's see how sampling works**

Let's consider a large population that follows a normal distribution, and we will take sample from it to study the sampling distribution of the sample mean.

```python
# create normal distribution with large sample size
normal_data = np.random.normal(15, 5, 1000)
# plot distribution
sns.histplot(normal_data)
```

create function which can generate provided number of sample with fixed sample size from the population and find mean of each sample and save it into list
- such as create 1000 samples where sample size is 30, find mean of each sample and save it into list

```python
# function
def sample_mean_function(popl, sample, sample_size):
    sample_mean = []   #empty list stores the output.

    for i in range(sample):
      sample2 = np.random.choice(popl,sample_size) #random size is being taken.
      mean = sample2.mean()   #provides statistical mean
      sample_mean.append(mean) #appends to the empty list

    sns.histplot(sample_mean)   #plots the histogram
    plt.axvline(np.mean(sample_mean), color = 'red') #add a vertical line for visibility.
    plt.title("sample mean plot") #provides plot title
    plt.xlabel("sample mean") #provides x axis label
    plt.ylabel("population")  #provides y axis label
    plt.show()    #shows the plot

```

```python
# run simulation with desire sample size and number of samples
sample_mean_function(random_list, int(input("provide number of sample:")), int(input("provide sample size:")))
```

## **What is Central limit theorem (CLT)?**

The central limit theorem represents that the sampling distribution of the mean will always be normally distributed, as long as the sample size is large enough.

For Better understanding let's take example of exam scores. suppose we have population of exam scores which is randomly generated so, we don't have any information on distribution.
